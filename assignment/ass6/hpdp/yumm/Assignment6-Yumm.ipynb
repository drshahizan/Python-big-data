### Load Less Data:
Strategically load only the essential portions of the dataset to optimize memory usage.


Drop columns that are unused

```python
columns_to_drop = ['l1', 'l2', 'l3', 'l4', 'l5', 'l6', 'rooms', 'bedrooms', 'bathrooms']

start_time = time.time()
df_less1 = pd.read_csv('ar_properties.csv', usecols=lambda column: column not in columns_to_drop)
end_time = time.time()

time_taken = end_time - start_time
print(f"Time taken to read the CSV file: {time_taken} seconds")
```

```python
df_less1.info()
```

```python
print('Missing values before: ')
df_less1.isna().sum()
```

Drop missing values in each column

```python
df_less1.dropna(subset=['lat'], inplace=True)
df_less1.dropna(subset=['lon'], inplace=True)
df_less1.dropna(subset=['surface_total'], inplace=True)
df_less1.dropna(subset=['surface_covered'], inplace=True)
df_less1.dropna(subset=['price'], inplace=True)
df_less1.dropna(subset=['currency'], inplace=True)
df_less1.dropna(subset=['price_period'], inplace=True)
df_less1.dropna(subset=['description'], inplace=True)
```

As we can see, the memory usage reduce from 190.7 MB (**df**) to 17.5 MB (**df_less1**)

```python
df_less1.info()
```

### Use Chunking:
Process the data in smaller pieces to avoid memory issues.


```python
start_time = time.time()

# Reading the data in chunks
data = pd.read_csv('ar_properties.csv', chunksize=300000)
dfs = []
for chunk in data:
    dfs.append(chunk)

# Concatenating the chunks together
df_less2 = pd.concat(dfs)

end_time = time.time()

time_taken = end_time - start_time
print(f"Time taken to read and concatenate the CSV file in chunks: {time_taken} seconds")
```

### Sampling :
Implement sampling methodologies to extract meaningful insights from a subset of the dataset.


##Random Sampling:
- Use only 100000 rows from the entire rows
- Equal probability of the selection by using random sampling


```python
file_path='/content/ar_properties.csv'
sample_size = 100000

# Random Sample

df_random_sample = pd.read_csv(file_path).sample(n=sample_size, random_state=42)
print("Random Sample:")
print(df_random_sample.head())

# Stratified Sample
df_stratified_sample = pd.read_csv(file_path)
stratified_sample = df_stratified_sample.groupby('ad_type', group_keys=False).apply(lambda x: x.sample(min(len(x), sample_size), random_state=42))
print("\nStratified Sample:")
print(stratified_sample.head())

# Time-Based Sample
df_time_based_sample = pd.read_csv(file_path)
df['created_on'] = pd.to_datetime(df['created_on'])
df_time_based_sample = df_time_based_sample.sort_values(by='created_on').head(sample_size)
print("\nTime-Based Sample:")
print(df_time_based_sample.head())

# Percentage-Based Sample
df_percentage_sample = pd.read_csv(file_path).sample(frac=0.1, random_state=42)
print("\nPercentage-Based Sample:")
print(df_percentage_sample.head())
```

```python
import psutil
import time
import os
import sys

memory_before_random = psutil.Process().memory_info().rss / 1024 / 1024
start_time_random = time.time()
df_random_sample = pd.read_csv(file_path).sample(n=sample_size, random_state=42)
end_time_random = time.time()
memory_after_random = psutil.Process().memory_info().rss / 1024 / 1024
time_taken_random = end_time_random - start_time_random
file_size_random = os.path.getsize(file_path) / (1024 * 1024)

print("Random Sample:")
print(df_random_sample.head())
print(f"Memory used: {memory_after_random - memory_before_random} MB")
print(f"Time taken: {time_taken_random} seconds")
print(f"File size: {file_size_random} MB")
print("\n---")
```

### Parallelize with Dask:
- Dask is a powerful library that extends pandas to enable parallel and distributed computing. It's particularly useful for handling larger-than-memory datasets.
- Scalability: Dask allows you to scale your computation from a single machine to a cluster of machines. This makes it suitable for handling large datasets and computationally intensive tasks that may not fit into the memory of a single machine.

- Parallel Processing: Dask enables parallel processing by breaking down a task into smaller tasks and executing them concurrently. This can significantly speed up the computation time, especially for tasks that can be parallelized.

- Lazy Evaluation: Dask uses lazy evaluation, meaning that it builds up a task graph without immediately executing the operations. This allows for more efficient task scheduling and resource allocation. The actual computation is triggered only when the results are explicitly requested, allowing for optimization opportunities.
