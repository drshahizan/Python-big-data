## 3.2. chunking

 Chunking involves reading data in smaller portions, or 'chunks', instead of loading the entire dataset at once.

 The Chunking process involves seperating data into smaller portionm, knows as chunks an it  will be repeated until the entire dataset has been splitted into chunks

 * the first step involves identifying how many data needed in a chunk, or chunk size
 * second, use he code below to start chunking

```python
import csv
import sys

chunk_size = 700000
num = 1
csv.field_size_limit(sys.maxsize)

with open(path) as file_obj :
    reader = csv.reader(file_obj)
    header = colnames

    while True :
      rows = [header]
      for _ in range(chunk_size):
        rows.append(next(reader))
        if len(rows) == 1:
          break;

      chunk_name = f'chunk{num}.csv'
      with open(chunk_name , 'w') as write_chunk:
        csv_writer = csv.writer(write_chunk)
        csv_writer.writerows(rows)
        num += 1
```

* the code above wil separate the entire dataset into chunks
* to view the data in each chunks, use the following code

```python
chunk_path = '/content/chunk31.csv'
df_chunk = pd.read_csv(chunk_path)
df_chunk.info(memory_usage='deep')
```

## 3.3. Data type optimization

Data type optimization is the process of changing one data type into its smaller form

example : int64 is the largest integer datatypes,therfore it can be change into smaller version such as int32, int16 or int8

we reduce the data type size to reduce memory and ram usage.

* The following codes shows the memory usage for 100000 rows of data before data type optimization

```python
df_optimize = pd.read_csv(path, nrows = 100000, header=None, names=colnames)
df_optimize.info(memory_usage='deep')
df_optimize
```

* As you can see, the data used 91.6 MB of memory with 1 field containing int64 datatype
<br><br>
* To change the datatype, use the following codes

```python
for column in df_optimize:
  if df_optimize[column].dtype == 'float64':
    df_optimize[column]=pd.to_numeric(df_optimize[column], downcast='float')
  if df_optimize[column].dtype == 'int64':
    df_optimize[column]=pd.to_numeric(df_optimize[column], downcast='integer')
```

* check the memory usage again

```python
df_optimize.info(memory_usage ='deep')
```

* as you can see, the memory usage has been reduce to 91.2 MB
