## 2. Chunking

Process the data in smaller pieces to avoid memory issues.
- Now, we are going to chunk the data with the size of 100,000 data per chunk.
- Chunking offers several advantages such as:
  - Memory efficiency. When working with large datasets that may not fit into memory entirely, chunking allows you to process smaller portions (chunks) of the data at a time.
  - Improved performance. Processing data in smaller chunks can lead to improved performance, as it reduces the need to load the entire dataset into memory.
  - Flexibility and scalability. Chunking provides flexibility in handling datasets of varying sizes. It allows you to scale your data processing approach based on the available resources. It is particularly useful when dealing with datasets that may grow over time.

```python
# Defining our desired chunk size for every chunks.
chunk_size = 100000
```

```python
# Calculate the number of chunks based on the DataFrame size
num_chunks = len(df) // chunk_size + 1

num_chunks
```

We should have 71 chunks of our dataset after performing the chunking process after this.

```python
# Iterate through chunks
for i in range(num_chunks):
    start_idx = i * chunk_size
    end_idx = (i + 1) * chunk_size if (i + 1) * chunk_size < len(df) else len(df)

    # Extract the chunk based on indices
    chunk = df.iloc[start_idx:end_idx]

    # For example, we print the first five rows of each chunk
    print(f"Chunk {i + 1}:")
    print(chunk.head())
    print("\n")
```

## 4. Sampling

- Implement sampling methodologies to extract meaningful insights from a subset of the dataset.

- Sampling allows you to extract a random subset of your data for analysis, which is useful for various purposes such as exploratory data analysis, model training on a smaller subset, or generating insights without processing the entire dataset.

#### Simple Random Sampling

```python
# Check unique values and their counts
# for the column 'Primary Type'
df['Primary Type'].value_counts()
```

```python
# Get ratio instead of raw numbers using normalize=True
expected_ratio = df['Primary Type'].value_counts(normalize=True)

# Round and then convert to percentage
expected_ratio = expected_ratio.round(4)*100

# convert to a DataFrame and store in variable 'primary_type_ratios'
# We'll use this variable to compare ratios for samples
# selected using SRS and Stratified Sampling
primary_type_ratios = pd.DataFrame({'Expected':expected_ratio})
primary_type_ratios
```

```python
# Choose a Simple Random Sample of 5 records
df.sample(n=20)
```

Or, we can use the parameter frac to randomly select a fraction of the dataset.

```python
# Choose an SRS with 20% of the dataset
srs_sample = df.sample(frac=0.20)
# Ratio of selected items by the Primary Type
srs_ratio = srs_sample['Primary Type'].value_counts(normalize=True)
# Convert to percentage
srs_ratio = srs_ratio.round(4)*100
# We did sampling using SRS. So give it proper name
srs_ratio.name = 'SRS'

# Finally add SRS sample proportions as a column to
# the  variable primary_type_ratio
primary_type_ratios = pd.concat([primary_type_ratios, srs_ratio], axis=1)
primary_type_ratios
```

#### Stratified Sampling

```python
# Stratified Sampling
# Use groupby and apply to select sample
# which maintains the population group ratios
stratified_sample = df.groupby('Primary Type').apply(lambda x: x.sample(frac=0.20))

stratified_sample.head()
```

```python
# Remove the extra index added by groupby()
stratified_sample = stratified_sample.droplevel(0)
stratified_sample.head()
```

Now, we calculate the proportion by each primary type for this sample.

```python
# Ratio of selected items by the primary type
stratified_ratio = stratified_sample['Primary Type'].value_counts(normalize=True)
# Convert to percentage
stratified_ratio = stratified_ratio.round(4)*100
# We did stratified sampling. So give it proper name
stratified_ratio.name = 'Stratified'

# Add it to the variable primary_type_ratios which already has
# the  expected and SRS proportions
primary_type_ratios = pd.concat([primary_type_ratios, stratified_ratio], axis=1)
primary_type_ratios
```

## 5. Parallelize with Dask

- Dask is a parallel computing library that integrates seamlessly with pandas, allowing for the parallel and distributed processing of large datasets.

-  Traditional data processing methods often struggle with large datasets that don't fit into memory. Thus Dask is an efficient way processing  Large Datasets

- The file size of our dataset '**Crimes_-_2001_to_Present.csv**' is 1.71GB which is pretty big

- Let see how Dask handle the datasets easy and quick

Install Dask and Import Dask to Read Data

```python
pip install dask
```

```python
import dask.dataframe as dd
import time

# Read the data using Dask without specifying data types
start_time = time.time()
# Due to mismatch detected by Dask, we state 'IUCR' to object
ddf = dd.read_csv('Crimes_-_2001_to_Present.csv', assume_missing=True, dtype={'IUCR': 'object'})
end_time = time.time()

# Drop rows with non-convertible values in specified columns
ddf = ddf.drop(columns=['IUCR'])
columns_to_check = ['Community Area', 'Ward']
for col in columns_to_check:
    ddf = ddf[ddf[col].notnull()]  # Drop rows with NaN values
    ddf = ddf[ddf[col].astype(str).str.isnumeric()]  # Drop rows with non-numeric values
```

Eventhough the datasets is 1.71 Gb, with Dask we can do it quickly

```python
print(f"Time taken to read data with Dask: {end_time - start_time} seconds")
```

According to the outcomes below, memory used by Dask to compute was 0, which is very efficient

```python
# Example: Compute the mean of the 'Year' column
start_time = time.time()
result = ddf['Year'].mean().compute()
end_time = time.time()

print(f"Time taken for Dask mean operation: {end_time - start_time} seconds")
print(f"Memory used by Dask: {ddf['Year'].memory_usage(deep=True).compute() / (1024**2):.2f} MB")
print(f"Mean value of 'Year' column using Dask: {result}")
```

- The code computes the mean of the 'Year' column to started in parallel and completed, which returns the result as a Pandas dataframe (result).

- With its ability to analyse enormous amounts of data in parallel, Dask is an invaluable tool for processing big amounts of data quickly.

## 6. Comparative Analysis of Big Data Processing Strategies


| Method                       | Memory Usage         | Computation Time    | File Size                | Advantages                                          | Disadvantages                                      |
|------------------------------|----------------------|---------------------|--------------------------|-----------------------------------------------------|----------------------------------------------------|
| Traditional Methods          | High                 | Longer              | Large                    | Straightforward but may struggle with large datasets | Limited scalability, potential memory overflow    |
| Load Less Data               | High (for loaded portion)| Short (for loaded portion)| Depends on loaded portion| Reduces memory load, suitable for limited analysis   | Limited insights due to partial dataset loading     |
| Use Chunking                 | Medium (depends on chunk)| Short (per chunk)    | Depends on chunk size     | Reduces memory load, suitable for larger datasets   | Complex implementation, potential data loss        |
| Optimize Data Types          | Medium to Low (after optimization)| Short          | Same as original file    | Efficient for reducing memory usage                  | May require careful handling of precision and scale|
| Sampling                     | Low (for sample)      | Short (for sample)  | Depends on sample size   | Provides insights with reduced computation          | Representativity challenges, potential bias        |
| Parallelize with Dask        | Low (chunk-wise)      | Short (parallelized)| Smaller file sizes       | Efficient for large-scale parallel processing       | Learning curve, potential overhead for small datasets|


- Memory usage, computation time, and file size will depend on various factors, including hardware, specific operations, and dataset characteristics.

- Advanced strategies, especially chunking and parallelization with Dask, offer advantages in handling larger-than-memory datasets, optimizing memory usage, reducing computation time, and resulting in smaller file sizes.
