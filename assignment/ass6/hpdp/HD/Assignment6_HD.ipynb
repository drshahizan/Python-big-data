## 4.  Strategies for Big Datasets



###1. Load less data

```python
selected_columns = ["DEP_DELAY", "ARR_DELAY", "TAXI_OUT","CRS_DEP_TIME", "DEP_TIME","CRS_ARR_TIME", "ARR_TIME", "AIRLINE_CODE"]
df_selected = pd.read_csv(file_path, usecols=selected_columns)
```

### 2. Process the data in smaller chunks

1.  This step allows you to read the dataset in smaller chunks rather than loading the entire dataset into memory at once. The *chunksize* parameter specifies the number of rows to be read in each iteration.

```python
chunk_size = 10000
chunks = pd.read_csv(file_path, usecols=selected_columns, chunksize=chunk_size)
```

2. Display the first few rows of the first chunk

```python
for chunk in chunks:
    print(chunk.head())
    break
```

- The code then iterates over the chunks using a *for* loop. In each iteration, it prints the first few rows of the current chunk using *chunk.head()*. The *break* statement is used to exit the loop after processing the first chunk. This is done for demonstration purposes to show the structure of the data in the first chunk.

###3. Optimize Data Types:


1. Fine-tune data types

```python
df_optimized = pd.read_csv(file_path,usecols=selected_columns, dtype={"DEP_DELAY" :"float32", "ARR_DELAY":"float32", "TAXI_OUT":"float32","CRS_DEP_TIME":"int32", "DEP_TIME":"float32","CRS_ARR_TIME":"int32", "ARR_TIME":"float32"})
```

- The key aspect here is the *dtype* parameter, which allows you to specify the data types for specific columns. In this case, it's explicitly stating that certain column should be of type *int32*, and certain should be of type *float32*.

2. Print the data types of each column

```python
df_optimized.dtypes
```

- Fine-tuning data types during the data loading phase is a strategy to optimize memory usage, enhance computational performance, and ensure compatibility with downstream analyses, ultimately contributing to more efficient and effective data processing.

### 5. Parallelize with Dask

1. Install Dask


```python
!pip install dask
```

- Dask is a parallel computing library that extends the capabilities of Pandas for handling larger-than-memory datasets.

2. Import Dask

```python
import dask.dataframe as dd
```

- This line imports the Dask dataframe module *dd* so that you can use Dask's dataframe functionalities. Dask dataframes are similar to Pandas dataframes but are designed to work with larger datasets that may not fit into memory.

3. Read the Dataset using Dask

```python
ddf = dd.read_csv(file_path)
```

- Dask dataframes can operate on larger-than-memory datasets by breaking them into smaller partitions that can be processed in parallel

4. Perform Operations in Parallel

```python
result = ddf.groupby('AIRLINE_CODE').DEP_DELAY.mean().compute()
result
```

- The code performs a groupby operation on the *AIRLINE_CODE* column of the Dask dataframe. It calculates the mean of the *DEP_DELAY* column within each group. The compute() function is then used to trigger the parallel execution of the computation and return the result as a Pandas dataframe (result).

- The grouping and aggregation operations are performed in parallel, making use of Dask's ability to distribute computations across multiple cores or even on a cluster for larger datasets.



> Dask can be used to read and process a large dataset in a distributed and parallelized manner, particularly when dealing with operations like grouping and aggregation. It's a powerful tool for efficiently handling big data that may not fit into memory using traditional Pandas workflows



## 5. Comparative Analysis

1. Install memory-profiler

```python
!pip install memory-profiler
```

- *Memory-profiler* is a tool used to profile and analyze the memory usage of Python code

2. Load the Memory Profiler Extension

```python
%load_ext memory_profiler
```

- The *%load_ext* magic command is used to load the extension, allowing you to use memory profiling capabilities

3. Profile Memory Usage

```python
%memit df = pd.read_csv(file_path)
print("Memory Usage:")
print(df.memory_usage().sum())
```

- *%memit* is a magic command that is used to profile the memory usage of a specific code block. In this case, it reads the CSV file specified by the *file_path* variable using Pandas *(pd.read_csv(file_path))* and assigns the result to the variable df.

- The memory usage of the resulting dataframe *(df)* is then printed using print*("Memory Usage:")* followed by print *(df.memory_usage().sum())*. This provides information about the total memory usage of the dataframe

4. Profile Computation Time

```python
%timeit df = pd.read_csv(file_path)
```

- *%timeit* is a magic command that is used to profile the execution time of a specific code block. In this case, it measures the time taken to execute the *pd.read_csv(file_path)* operation.

- The result is printed automatically by *%timeit* and includes information such as the average time taken for the operation to complete.



> The *memory-profiler* and *%timeit* magic commands to profile the memory usage and computation time, respectively, of the operation of reading a CSV file into a Pandas dataframe. These tools are useful for identifying potential bottlenecks and optimizing code for better performance
