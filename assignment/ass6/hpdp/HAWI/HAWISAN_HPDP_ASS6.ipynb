# CHUNKING
- Iterates through chunks of the CSV file using pd.read_csv with the specified chunk_size. <br>
For each chunk: <br>
- Measures memory usage before and after processing the chunk using psutil.
- Records the start time and end time to calculate the computation time.
- Prints information about the current chunk, including shape, file size, memory usage, and computation time.
- Saves the processed chunk to a new CSV file.

```python
import pandas as pd
import os
import psutil
from time import time

cleaned_file_path = r"clean_202304.csv"

chunk_size = 10000
for i, chunk in enumerate(pd.read_csv(cleaned_file_path, chunksize=chunk_size)):

    memory_before = psutil.Process().memory_info().rss

    start_time = time()

    print(f"Chunk {i+1} Shape: {chunk.shape}")


    if i == 0:
        print(f"\nContents of Chunk {i+1}:\n{chunk}")


    chunk.to_csv(f"{cleaned_file_path}_chunk_{i+1}.csv", index=False)

    end_time = time()
    memory_after = psutil.Process().memory_info().rss
    computation_time = end_time - start_time

    file_size = os.path.getsize(f"{cleaned_file_path}_chunk_{i+1}.csv")
    print(f"Chunk {i+1} File Size: {file_size} bytes")
    print(f"Chunk {i+1} Memory Usage: {memory_after - memory_before} bytes")
    print(f"Chunk {i+1} Computation Time: {computation_time} seconds")

    if i == 0:
        print(f"\nContents of Chunk {i+1}:\n{chunk}")
```

#SAMPLING
- Initializes variables to measure memory usage before and after the sampling process and record the start time for computation time.
- Iterates through each row of the CSV file using pd.read_csv with chunksize=1.
- For the first n rows, appends each row to the sampled_rows list.
- For subsequent rows (beyond the first n), calculates the probability of replacing an existing sampled row with the new row based on a uniform distribution.
- If the new row is selected, a randomly chosen existing row is replaced in the sampled_rows list.
- Calculates memory usage after the sampling process and the total computation time.

```python
n = 10000

memory_before = psutil.Process().memory_info().rss
start_time = time()

sampled_rows = []
for i, row in enumerate(pd.read_csv('clean_202304.csv', chunksize=1)):
    if i < n:
        sampled_rows.append(row)
    else:
        probability = n / (i + 1)
        if random.random() < probability:
            replace_index = random.randint(0, n - 1)
            sampled_rows[replace_index] = row

memory_after = psutil.Process().memory_info().rss
computation_time = time() - start_time

sampled_df = pd.concat(sampled_rows, axis=0, ignore_index=True)

print("Sampled DataFrame:")
print(sampled_df)

print(f"Memory Usage: {memory_after - memory_before} bytes")
print(f"Computation Time: {computation_time} seconds")
```

# PARALLELIZE WITH DASK
- Uses dask.dataframe.read_csv to create a Dask DataFrame (dask_df) from the CSV file. Dask allows parallelized computation on larger-than-memory datasets.
- Measures memory usage before the computation using psutil.
- Records the start time for computation time measurement.
- Performs a groupby operation on the 'Property_Type' column, calculating the mean of the 'price' column using Dask's parallelized computing capabilities.
- Uses the compute() method to trigger the computation and retrieve the result as a pandas Series (dask_result).
- Records the end time for computation time measurement.
- Measures memory usage after the computation using psutil.

```python
import dask.dataframe as dd
import psutil
from time import time

dask_df = dd.read_csv('clean_202304.csv')

memory_before = psutil.Process().memory_info().rss

start_time = time()

dask_result = dask_df.groupby('Property_Type')['price'].mean().compute()

end_time = time()
memory_after = psutil.Process().memory_info().rss
computation_time = end_time - start_time

print("Dask Result:")
print(dask_result)

print(f"Memory Usage: {memory_after - memory_before} bytes")
print(f"Computation Time: {computation_time} seconds")
```

# OPTIMIZE DATA TYPES

- Uses dask.dataframe.read_csv to create a Dask DataFrame (dask_df) from the CSV file.
- Converts specific columns to more memory-efficient data types to optimize memory usage:
'price' to 'float32'.
'Date_of_Transfer' and categorical columns to 'category'.
- Measures memory usage before the fine-tuning process using psutil.
- Records the start time for computation time measurement (though there is no significant computation in this section).
- Records the end time for computation time measurement.
- Measures memory usage after the fine-tuning process using psutil.
- Calculates the total computation time for the fine-tuning process.

```python
import dask.dataframe as dd
import psutil
from time import time


dask_df = dd.read_csv('clean_202304.csv')

dask_df['price'] = dask_df['price'].astype('float32')
dask_df['Date_of_Transfer'] = dask_df['Date_of_Transfer'].astype('category')
dask_df['Property_Type'] = dask_df['Property_Type'].astype('category')
dask_df['Old/New'] = dask_df['Old/New'].astype('category')
dask_df['postcode'] = dask_df['postcode'].astype('category')
dask_df['Duration'] = dask_df['Duration'].astype('category')
dask_df['PAON'] = dask_df['PAON'].astype('category')
dask_df['SAON'] = dask_df['SAON'].astype('category')
dask_df['Locality'] = dask_df['Locality'].astype('category')
dask_df['Town/City'] = dask_df['Town/City'].astype('category')
dask_df['District'] = dask_df['District'].astype('category')
dask_df['County'] = dask_df['County'].astype('category')
dask_df['PPDCategory_Type'] = dask_df['PPDCategory_Type'].astype('category')

memory_before_processing = psutil.Process().memory_info().rss

start_time_processing = time()

end_time_processing = time()
memory_after_processing = psutil.Process().memory_info().rss
computation_time_processing = end_time_processing - start_time_processing

print(f"Memory Usage (After Fine-Tuning): {memory_after_processing - memory_before_processing} bytes")
print(f"Computation Time (After Fine-Tuning): {computation_time_processing} seconds")

fine_tuned_data_types = dask_df.dtypes
print("\nFine-Tuned Data Types:")
print(fine_tuned_data_types)
```

### Comparison Between Five Strategies

#### In Terms of Approach:
| Strategy                | Traditional Approach | Advanced Approach      
|-------------------------|------------------|-----------------------|
| Load Less Data          | Loads entire dataset into memory thus very memory intensive and less optimised | Only most relevant data are selected to load into memory |
| Chunking                | Processes all data in one go | Breaks data into smaller, manageable parts | Memory-efficient processing in smaller pieces |
| Data Type Optimization  | Automatically infers data types | Fine-tunes data types for efficiency (e.g. data type size)|
| Sampling                | Analyzes the entire dataset | Analyzes a representative subset (e.g., randomly or percentile of dataset) |
| Parallelize processing using the Dask library  | Lacks support for parallel computing | Uses Dask for parallel and distributed computing	|

#### In Terms of Pros and Cons:
| Strategy                | Pros | Cons
|-------------------------|----------|-------|
| Load Less Data          | Minimising memory usage, greatly improving handling of large data, making it possible to analyze big datasets | Limits the amount of data available for information gathering |
| Chunking                | Memory-efficient processing in smaller pieces | Some operations may require additional handling across chunks |
| Data Type Optimization  | Reduces memory usage significantly by using more compact data types | Careful selection of columns is crucial based on data distribution |
| Sampling                | Speeds up analysis, especially when a full-scale examination is not needed, can also be used to quickly analyse random or representative subset of dataset | May miss specific patterns or cannot capture rare occurrences present in the entire dataset |
| Parallelize with Dask   | Drastically cuts down computation time, enables working with datasets larger than memory | Requires Dask installation and adjustment of computations for parallelism |


# CONCLUSION


In conclusion, the choice of strategy in working with big data significantly influences the efficiency, resource utilization, and overall success of data analysis. Each strategy comes with its own set of benefits and drawbacks.

1. **Load Less Data**  
    
    Stands out for its emphasis on minimizing memory usage, thereby enhancing the handling of large datasets. However, it comes at the cost of limiting the amount of data available for analysis, potentially sacrificing comprehensive insights.

2. **Chunking**
    
    Addresses memory challenges by breaking down the data into manageable parts. While this offers memory-efficient processing, it introduces complexities in handling operations across chunks, which may require additional considerations.

3. **Data Type Optimization**
    
    Advantageous for reducing memory usage by employing more compact data types. However, careful column selection based on data distribution is crucial for its success.

4. **Sampling**
    
    Accelerates analysis by examining a representative subset of the dataset, but it runs the risk of missing specific patterns or failing to capture rare occurrences present in the entire dataset.

5. **Parallelize with Dask**
    
    Drastically cuts down computation time and enabling the analysis of datasets larger than memory. However, it demands the installation of Dask and adjustments to computations for parallelism.

In summary, the selection of a strategy should align with the specific requirements of the analysis, considering factors such as data size, memory constraints, and the need for comprehensive insights. An informed choice based on these considerations will contribute to more efficient and effective big data analysis.
