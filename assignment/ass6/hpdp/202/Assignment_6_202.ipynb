##Strategies for Big Datasets
Clearly explain the steps for each strategy. Help us understand how to make these strategies work.

###1. Load Less Data
In this strategy, we only select and load the useful columns from dataset for analysis.

* In this example, we only load columns: 'OP_CARRIER', 'ORIGIN', 'DEST', 'DEP_TIME', 'DEP_DELAY', 'ARR_TIME','ARR_DELAY','DISTANCE' and 'FL_DATE' for analysis. Add a new list to select the wanted columns.

* While reading the dataset, add a 'usecols' argument within the read_csv() function to read specific columns from a CSV file into a pandas DataFrame.

```python
selected_columns = ['OP_CARRIER', 'ORIGIN', 'DEST', 'DEP_TIME', 'DEP_DELAY', 'ARR_TIME','ARR_DELAY','DISTANCE', 'FL_DATE']

df_lessData = pd.read_csv('airline delay analysis/2017.csv', usecols=selected_columns)
df_lessData
```

###2. Use Chunking
In chunking, we split the file into smaller pieces.

* First, we decide how big the size of the chuck we want. In this case, we choose 5000 rows.
* After that, we load the dataset and set the chuck size as the size desired. The loop will then chunk the dataset and rename it as 'chunk1', 'chuck2' etc.

```python
chunk_size = 5000
num = 1
result_df = pd.DataFrame()

path = 'airline delay analysis/2017.csv'

chunks = pd.read_csv(path, chunksize=chunk_size)

for chunk in chunks:
    chunk.to_csv('chunk'+str(num)+'.csv', index=False)
    gc.collect()
    num+=1
```

* Checking whether the chucking is successful.

```python
df_chuck = pd.read_csv('chunk25.csv')
df_chuck
```

###3. Optimize Data Types

### b. Advantages and Disadvantages of Strategies

#### Traditional Approach

**Load Entire Dataset:**

* Memory Usage: High

* Computation Time: Slower due to memory constraints.

* File Size: The bigger the dataset size, the bigger the file size.

#### Advanced Approaches

**Load Less Data:**

* Memory Usage: Reduced by only loading columns needed.

* Computation Time: Faster as less data to process.

* File Size: Unchanged.

**Use Chunking:**

* Memory Usage: Lower as data is in chunks.
* Computation Time: Takes more time to process the data compare to other approach.
* File Size: Unchanged.

**Optimize Data Types:**

* Memory Usage: Reduced by reducing the size of data types.
* Computation Time: Faster due to less memory used.
* File Size: Unchanged.

**Sampling:**

* Memory Usage: Reduced by loading a sample of the data.
* Computation Time: Faster due to smaller dataset size, but is affected by the size of sampled data.
* File Size: Reduced, but may not represent the entire dataset.

**Parallelize with Dask:**

* Memory Usage: Efficiently handles larger-than-memory datasets by parallelizing operations.
* Computation Time: Fastest compared to other approches.
* File Size: Unchanged.

##Conclusion

In conclusion, advanvced strategies is much more effiecient compared to traditional approach. In terms of memory efficiency, traditional approach often face challenges with high memory usage and sometimes the connection is lost due to insufficient memory, limiting the ability to process large datasets. By using advanced strategies, we can enhance the memory efficiency by focusing on loading only essential data, processing in chunks, and optimizing data types. This enables us to handle large datasets that may exceed available memory.

In terms of computational time, traditional approach tend to be slower due to memory constraints and it needs to process the dataset in a single pass. By using advanced strategies, the computation times is faster through parallel processing, efficient memory usage, and loading only targeted data. Parallelize with Dask is the most efficient in computational time.

Using these advanced strategies, we can process large dataset even when we only have limited memory space available. The computational time can be reduced too.
