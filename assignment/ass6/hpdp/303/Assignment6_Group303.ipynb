Chooses specific columns (like "CARRIER_NAME," "DEPARTING_AIRPORT," etc.) from a dataset that contains information about airline delays and cancellations. Loads only the selected columns from a large dataset (CSV file) into a smaller DataFrame using the Pandas library.

#### Use Chunking: Process the data in smaller pieces to avoid memory issues.


This code reads a large CSV file in chunks of 10000 rows each, and saves each chunk to a separate CSV file named 'chunk1.csv', 'chunk2.csv', and so on. It's a memory-efficient way to handle large datasets. The garbage collector is used to manage memory during the process.

```python
import gc

chunk_size = 10000
num = 1
chunk_result_df = pd.DataFrame()
total_memory_usage = 0

file_path = '/content/2019-airline-delays-and-cancellations/full_data_flightdelay.csv'

chunks = pd.read_csv(file_path, chunksize=chunk_size)

# Start measuring time
chunk_start_time = timeit.default_timer()

for chunk in chunks:
    if num <= 5:
      print(f"Chunk {num}:\n", chunk.head())
    chunk_memory_usage = chunk.memory_usage(deep=True).sum() / (1024 ** 2)
    total_memory_usage += chunk_memory_usage
    chunk.to_csv('chunk'+str(num)+'.csv', index=False)
    gc.collect()
    num+=1

# Stop measuring time
chunk_end_time = timeit.default_timer()
chunk_elapsed_time = chunk_end_time - chunk_start_time
```

```python
print(f"\nTotal memory usage for all chunks: {total_memory_usage:.2f} MB")
print(f"Total elapsed time for processing chunks: {chunk_elapsed_time:.2f} seconds")
```

#### Optimize Data Types: Fine-tune data types to maximize efficiency and minimize memory consumption.


Converting specific data types to more memory-efficient alternatives, and displays data types and memory usage before and after optimization. The optimization involves converting float64 to float16, int64 to int16, object to category.

```python
# Display data types and memory usage before optimization
print("Data types before optimization:")
print(df.dtypes)
print("\nMemory usage before optimization:")
print(df.memory_usage(deep=True).sum() / (1024 * 1024), "MB")

# Optimize data types
for col in df.columns:
    if df[col].dtype == 'float64':
        df[col] = df[col].astype('float16')
    if df[col].dtype == 'int64':
        df[col] = df[col].astype('int16')
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category')

# Display data types and memory usage after optimization
print("\nData types after optimization:")
print(df.dtypes)
print("\nMemory usage after optimization:")
print(df.memory_usage(deep=True).sum() / (1024 * 1024), "MB")
```

```python
print("\nMemory usage for optimized data types:", df.memory_usage(deep=True).sum() / (1024 * 1024), "MB")

# Measure computation time on the optimized DataFrame
import timeit
time = timeit.timeit(lambda: df, number=1)
print("Computation time for optimize data types:", time, "seconds")
```

#### Parallelize with Dask: Dask is a powerful library that extends pandas to enable parallel and distributed computing. It's particularly useful for handling larger-than-memory datasets.

Converts a Pandas DataFrame (df) into a Dask DataFrame (dask_df) with 10 partitions. Computes the mean of the 'DISTANCE_GROUP' column using Dask. Executes the computation and prints the computed mean distance. Performs a parallelized groupby operation on 'DEP_TIME_BLK' and calculates the sum of 'NUMBER_OF_SEATS' for each group.

```python
!pip install dask
```

```python
import dask
import dask.dataframe as dd

# Assuming 'df' is your Pandas DataFrame
dask_df = dd.from_pandas(df, npartitions=10)  # Adjust the number of partitions as needed

# Calculate the mean of a column
mean_distance = dask_df['DISTANCE_GROUP'].mean()

# Compute the result
result_mean_distance = mean_distance.compute()
print("Mean Distance:", result_mean_distance)

# Perform a parallelized groupby operation
result_groupby = dask_df.groupby('DEP_TIME_BLK')['NUMBER_OF_SEATS'].sum().compute()
print("\nGrouped by DEP_TIME_BLK, Sum of NUMBER_OF_SEATS:")
print(result_groupby)
```

```python
# Measure memory usage for the Dask DataFrame
print("Memory usage for parallelize with Dask:", dask_df.memory_usage(deep=True).sum().compute() / (1024 ** 2), "MB")

# Measure computation time on the Dask DataFrame
import timeit
time = timeit.timeit(lambda: dask_df.compute(), number=1)
print("Computation time for parallelize with Dask:", time, "seconds")
```
