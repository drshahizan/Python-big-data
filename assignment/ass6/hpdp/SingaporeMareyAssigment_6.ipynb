# **2. Use chunking**

**Step 1**: Determine an appropriate chunk size based on the available memory and the size of the dataset.

**Step 2**: Use the chunksize parameter in Pandas' read_csv function to read the dataset in chunks.

**Step 3**: Iterate through the chunks using a for loop.

**Step 4**: Process each chunk independently, potentially saving results after each iteration.

```python
import pandas as pd
import time
import os

# Define chunk size
chunk_size = 100000

# Display memory usage, computation time, and file size for each chunk
for chunk_number, chunk in enumerate(pd.read_csv('male_players.csv', chunksize=chunk_size), start=1):
    print(f"\nProcessing Chunk {chunk_number}:")

    # Display memory usage for the current chunk
    print("Memory Usage:")
    chunk.info(memory_usage='deep')

    # Measure computation time for the current chunk
    start_time = time.time()

    # Perform your data analysis or processing on the current chunk
    # Example: Calculate the average overall rating for each chunk
    avg_overall = chunk['overall'].mean()
    print(f"Average Overall Rating for Chunk {chunk_number}: {avg_overall}")

    end_time = time.time()
    computation_time = end_time - start_time
    print(f"Computation Time: {computation_time} seconds")

# File Size (total for the entire dataset)
file_size = os.path.getsize('male_players.csv') / (1024 ** 2)  # Convert to MB
print(f"\nFile Size (Full Dataset): {file_size} MB")


```

# **4. Sampling**


**Step 1**: Determine the desired sampling percentage based on your analysis needs.

**Step 2**: Use the 'nrows' method in Pandas to randomly select a subset of rows.

**Step 3**: Optionally, set the random_state parameter for reproducibility.

**Step 4**: Perform your analysis on the sampled dataset.

```python
import pandas as pd
import time
import os

# Measure memory usage before loading the dataset
start_memory = os.popen('free -t -m').readlines()
start_memory = int(start_memory[1].split()[1])

# Measure computation time
start_time = time.time()

# Load the dataset
df_full = pd.read_csv('male_players.csv', nrows = 100000)

# Create a random sample with 20% of the dataset
sampled_df = df_full.sample(frac=0.5, random_state=42)

# Display memory usage before sampling
print("Memory Usage (Before Sampling):")
df_full.info(memory_usage='deep')

# Perform your data analysis on the entire dataset if needed
# Example: Calculate the average overall rating for the full dataset
avg_overall_full = df_full['overall'].mean()
print(f"\nAverage Overall Rating for the Full Dataset: {avg_overall_full}")

# Display memory usage after sampling
print("\nMemory Usage (After Sampling):")
sampled_df.info(memory_usage='deep')

# Measure computation time
end_time = time.time()
computation_time = end_time - start_time
print(f"\nComputation Time: {computation_time} seconds")

# Measure memory usage after loading the dataset
end_memory = os.popen('free -t -m').readlines()
end_memory = int(end_memory[1].split()[1])

# File Size (total for the entire dataset, not affected by sampling)
file_size = os.path.getsize('male_players.csv') / (1024 ** 2)  # Convert to MB
print(f"\nFile Size (Full Dataset): {file_size} MB")
```

#**Comparison for each strategies applied**

**Load Less Data**:
Memory Usage: Significantly reduced by loading only a subset of the data, optimizing memory consumption.
Computation Time: Generally faster due to the reduced amount of data being processed.
File Size: Unaffected, as only a subset is loaded.

**Chunking**:
Memory Usage: Managed effectively by processing the data in smaller, more manageable chunks. Reduces the risk of running out of memory.
Computation Time: May increase slightly due to the overhead of processing multiple chunks, but can be more scalable.
File Size: Unaffected, as the chunks are processed independently.

**Optimize Data Types**:
Memory Usage: Reduced by converting data types to more memory-efficient ones, especially numeric types.
Computation Time: May improve due to smaller memory footprint, allowing for faster calculations.
File Size: Remains the same, but memory efficiency is increased.

**Sampling**:
Memory Usage: Reduced as you are working with a smaller subset of the data.
Computation Time: Faster, as you are analyzing a smaller portion of the dataset.
File Size: Unaffected, as the original dataset size remains constant.

**Parallelize with Dask:**
Memory Usage: Managed efficiently through parallel processing, allowing for larger-than-memory datasets.
Computation Time: Potentially faster due to parallelization, especially for computationally intensive tasks.
File Size: Unaffected, but loading is distributed across multiple cores or machines.

#**Comparative Analysis**:

**Memory Usage**: Advanced strategies (Load Less Data, Chunking, Optimize Data Types, Sampling, Parallelize with Dask) generally result in significantly reduced memory usage compared to the traditional method, which loads the entire dataset into memory.

**Computation Time**: Advanced strategies can lead to faster computation times by processing data more efficiently, though the impact may vary based on the specific strategy and dataset characteristics.

**File Size**: Traditional and advanced strategies do not affect the original file size, except for Load Less Data, which deals with a subset of the data.

#**Conclusion**:

The choice of strategy depends on the specific requirements and constraints of the data processing task. For large datasets, traditional methods may not be feasible due to memory constraints. Advanced strategies provide efficient alternatives, each with its advantages. Load Less Data, Chunking, and Optimizing Data Types are effective for managing memory, while Sampling and Parallelizing with Dask offer scalable solutions for computation. The selection depends on the nature of the analysis, available resources, and the desired trade-offs between memory usage, computation time, and scalability. Adopting advanced strategies is crucial for handling big data effectively and avoiding performance bottlenecks associated with traditional methods.
