###Use Chunking:
Process the data in smaller pieces to avoid memory issues.

In order to concat for all 3 csv files, the chunking technique has been used, the original file unable to merge all the 3 csv files as it will cause system RAM clash.

*   Processing Time is 224.17 seconds.
*   System RAM Usage is 7.40% / 13.1 MB

```python
import pandas as pd

# Define chunk size
chunk_size = 10000

# Create iterators for each CSV file
games_iter = pd.read_csv('/content/games.csv', chunksize=chunk_size)
rec_iter = pd.read_csv('/content/recommendations.csv', usecols=['app_id', 'helpful', 'funny', 'date', 'is_recommended', 'hours', 'user_id', 'review_id'], chunksize=chunk_size)
users_iter = pd.read_csv('/content/users.csv', usecols=['user_id', 'products', 'reviews'], chunksize=chunk_size)

# Initialize lists to store processed chunks
games_chunks = []
rec_chunks = []
users_chunks = []

# Loop through chunks for each CSV file
for games_chunk, rec_chunk, users_chunk in zip(games_iter, rec_iter, users_iter):
    # Merge with other DataFrames if needed
    # Drop duplicate columns or perform other processing steps

    # Append processed chunks to the lists
    games_chunks.append(games_chunk)
    rec_chunks.append(rec_chunk)
    users_chunks.append(users_chunk)

# Concatenate all processed chunks for each CSV file
final_games = pd.concat(games_chunks, ignore_index=True)
final_rec = pd.concat(rec_chunks, ignore_index=True)
final_users = pd.concat(users_chunks, ignore_index=True)

# Concatenate the three DataFrames vertically
final_combined = pd.concat([final_games, final_rec, final_users], axis=0, ignore_index=True)

# Calculate processing time in seconds
end_time = time.time()
total_time_seconds = end_time - start_time
print(f"Processing Time: {total_time_seconds:.2f} seconds")

# Get RAM usage percentage
ram_percentage = psutil.virtual_memory().percent
print(f"System RAM Usage: {ram_percentage:.2f}%")

# Check the final_combined data size
final_games.info(memory_usage='deep')
```

5. Users DataFrame (Full):

Overall, the chunked approach allows you to process the data in smaller portions, reducing memory usage. However, it may require additional processing steps to handle chunks individually. The entire dataset approach loads the complete dataset into memory, which can be faster but may consume more resources.

```python
import pandas as pd

# Define chunk size
chunk_size = 10000

# Read the datasets using chunks
games_df = pd.read_csv('/content/games.csv', chunksize=chunk_size)
rec_df = pd.read_csv('/content/recommendations.csv', chunksize=chunk_size)
users_df = pd.read_csv('/content/users.csv', usecols=['user_id', 'products', 'reviews'], chunksize=chunk_size)

# Method 1: Using Chunks
def process_chunks():
    final_combined_chunks = pd.DataFrame()

    for games_chunk, rec_chunk, users_chunk in zip(games_df, rec_df, users_df):
        # Perform operations on each chunk
        combined_chunk = pd.merge(games_chunk, rec_chunk, on='app_id', how='left')
        combined_chunk = pd.merge(combined_chunk, users_chunk, on='user_id', how='left')

        # Concatenate chunks
        final_combined_chunks = pd.concat([final_combined_chunks, combined_chunk], ignore_index=True)

    return final_combined_chunks

# Method 2: Processing Entire Dataset by using Pandas (without merge)
def process_entire_dataset():
    # Read the entire datasets
    games_df_full = pd.read_csv('/content/games.csv')
    rec_df_full = pd.read_csv('/content/recommendations.csv')
    users_df_full = pd.read_csv('/content/users.csv', usecols=['user_id', 'products', 'reviews'])

    return games_df_full, rec_df_full, users_df_full

# Measure resources and processing time for chunks
final_combined_chunks = process_chunks()
print("Resources for final_combined_chunks:")
print(final_combined_chunks.info(memory_usage='deep'))
print("\n")

# Measure resources and processing time for entire dataset by using Pandas
start_time_entire = time.time()
games_df_full, rec_df_full, users_df_full = process_entire_dataset()
end_time_entire = time.time()
print(f"Processing time for entire dataset: {end_time_entire - start_time_entire} seconds")

# Display the first few rows of each DataFrame
print("Games DataFrame (Full):")
print(games_df_full.head())
print("\n")

print("Recommendations DataFrame (Full):")
print(rec_df_full.head())
print("\n")

print("Users DataFrame (Full):")
print(users_df_full.head())
```

##Conclusion

In conclusion, the comparative analysis between different data processing strategies provides valuable insights into the trade-offs and benefits associated with each approach. The original data processing method, while simple and intuitive, may exhibit limitations in terms of memory usage and processing time, particularly when dealing with large datasets.

The choice of which strategy performs better depends on the specific analytical goals, dataset characteristics, and available resources. For large datasets, strategies like chunking, optimizing data types, and parallelizing with Dask tend to offer better performance in terms of reduced memory usage and faster processing times. However, for smaller datasets or quick exploratory analysis, simpler approaches like loading less data or using the original processing method may suffice. A combination of strategies may often yield the best results in balancing performance and resource efficiency.
