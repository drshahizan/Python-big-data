The App Store is an app marketplace developed and maintained by Apple Inc., for mobile apps on its iOS and iPadOS operating systems. The store allows users to browse and download approved apps developed within Apple's iOS Software Development Kit. This dataset was found from Kaggle and its size is **406.04 MB**.

src: https://www.kaggle.com/datasets/gauthamp10/apple-appstore-apps

The dataset contains 123037 records and 21 columns which are:

1.   **App_Id** : Application ID
2.   **App_Name** : Application Name
3.   **AppStore_Url** : url link of the application
4.   **Primary_Genre** : category of the app falls in
5.   **Content_Rating** : age rating for the apps; 4+,18+,etc            
6.   **Size_Bytes** : storage size of the app
7.   **Required_IOS_Version** : Required IOS version to download app through appstore
8.   **Released** : yyyy-mm-dd hh:mm:ss format;date released                  
9.   **Updated**  : yyyy-mm-dd hh:mm:ss format;date updated                
10.  **Version** : version of the app                    
11.  **Price** : 0.00 format; Price to pay for the app
12.  **Currency** : currency of the app price                  
13.  **Free** : True=Free; False=Not Free                        
14.  **DeveloperId** : Developer ID                
15.  **Developer** : The deceloper name                 
16.  **Developer_Url** : link to the developer            
17.  **Developer_Website** : webiste of the developer        
18.  **Average_User_Rating** : average user rating of the app      
19.  **Reviews** : count of reviews                  
20.  **Current_Version_Score** : current version score    
21.  **Current_Version_Reviews**  : current version reviews.

## Exploratory Analysis and Visualization





Let's begin by importing matplotlib.pyplot and seaborn.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
df['Currency'].unique()
```

**Finding Mean**

```python
avg_sizebytes_genre = df.groupby('Primary_Genre').apply(lambda x: x['Size_Bytes'].mean()).reset_index(name= "Average Size_Bytes")
pd.set_option('display.precision',2)
avg_sizebytes_genre
```

Primary genre of book have the highest average of byte size.

**Barchart**

```python
primary_genre = df.groupby('Primary_Genre').apply(lambda x: len(x['App_Id'])).reset_index(name='Num_AppId')

fig, ax = plt.subplots(figsize=(30, 15))
sns.barplot(x=primary_genre['Primary_Genre'], y=primary_genre['Num_AppId'])
ax.set(xlabel ="Primary_Genre", ylabel = "Number of Apps ID", title ='Number of Apps by Genre')
plt.xticks(rotation=90)
plt.show()
```

Games is the highest primary genre in the Apple AppStore apps.

**Scatterplot**

```python
x = df['Average_User_Rating']
y = df['Price']

plt.title("Price by Average user rating")
plt.ylabel('Price')
plt.xlabel('Average User Rating')
plt.scatter(x, y)
plt.show()
```

Most expensive apps got no review.

## Asking and Answering Questions


#### Q1: Which app got the longest time between its released and updated time?

```python
df['Difference_Released_Updated'] = (df['Updated'] - df['Released']).dt.days
```

```python
df.rename(columns = {'Difference_Released_Updated':'Difference_Released_Updated(days)'}, inplace = True)
```

```python
df.head(2)
```

```python
df['Difference_Released_Updated(days)'].sort_values(ascending=False)
```

```python
df.iloc[69877]
```

**ANS1:** 'Map My Ride by Under Armour' app got the longest time to be updated from its release date.

#### Q2: Which developer produce the most apps?

```python
developer_most_apps.sort_values("Developer", ascending=False)
```
