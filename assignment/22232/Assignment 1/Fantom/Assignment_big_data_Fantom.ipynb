Double check if we have removed all null rows.

```python
print(df2.isnull().sum())
```

```python
def reduce_mem_usage(df):
    start_mem = df.memory_usage().sum() / 1024**3
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            df[col] = df[col].astype('category')

    end_mem = df.memory_usage().sum() / 1024**3
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df
```

```python
df3 = reduce_mem_usage(df2)
```

```python
df3.info(memory_usage='deep')
```

From the result of **df** and **df2** info(), we can observe that there are significant decrease in memory usage changing from 2.4GB to 664.2MB by loading the column that we are interested and alter the datatype of the columns.

## Asking and Answering Questions
A total of 5 questions will be asked in this section and all of them were answered by using pandas, series or visualization(matplotlib & seaborn).

### Q1: Which app have the most install in Google Playstore?

For this question we will use Pandas groupby() function to group the data according to the app name and sum the maximum install together and display in a bar chart.

First we group the app together by using groupby function and assign it into a variable name group_app.

```python
group_app = df3.groupby(['App Name'])['Maximum Installs'].sum()
group_app
```

```python
group_app.nlargest(10).plot(kind='bar', figsize=(10,5), color = 'blue')
plt.title("Top 10 Apps")
plt.ylabel('Number of Installs')
plt.xlabel('App Name')
```

From the barchart above we can observe that the Google Play Services have the most install followed by YouTube and Google. <br>
Due to the hectic lifestyle, people now a days tend to play games release their stress and Google Play Service is a must to save the game progress in cloud. Hence it have highest installation records.

### Q2: What is the percentage of free app for each Content Rating?
The function value_count() will be used to obtain the content occurance and the result will be display in a pie chart.

First we count the sum of number of content rating in Google Playstore.

```python
free_app = df3.loc[df3['Free'] == 1]
free_app_content = free_app['Content Rating'].value_counts()
free_app_content
```

```python
plt.figure(figsize=(20,10))

plt.pie(free_app_content, labels = free_app_content.index, autopct='%.0f%%') #Plot based on percentages
plt.title("Free app according to content rating")
```

Based on the pie chart above, it is very obvious that the "Everyone" content rating have the highest percentage compare to other content rating. 
Most of the free app required great install to boost their popularity. Therefore it is suggested to have content rating cover all age of people. 

### Q3: Which month have the most app being developed?

For this question we need to use dt.month function and groupby() function to count and group the app according to the month. The result will be display in a line plot.

```python
## First we calculate the sum of app group by month
month_app = df3.groupby(df3['Released'].dt.month).count()
```

```python
## Next we calculate the percentage according to each month
month_app['Percentage'] = (month_app['App Name'] / month_app['App Name'].sum()) * 100
```

```python
## Plotting the Lineplot
matplotlib.rc_file_defaults()
sns.set_style(style = None, rc = None )

plt.subplots(figsize = (10,5))
sns.lineplot(data = month_app['Percentage'], marker = 'o', sort = False, label = "Percentage Count")
```

From the line plot above we can conlcude that during the 5th month have the most app released by the developer. 

### Q4: Which category have the highest average price among premium app?

This question require groupby() function and mean to calculate the average price for each category.

```python
## First we select the app with price labeled.
premium_app = df3.loc[df3['Free'] == 0]

## Then we group by category and use the mean function
premium_app_group = premium_app.groupby(['Category'])['Price'].mean()
```

```python
## Plot a bar chart
premium_app_group.plot(kind='bar', figsize=(12,5), color = 'blue')
plt.title("Category with Highest Average Price")
plt.ylabel('Average Price')
plt.xlabel('Category')
```

From the bar chart above we can observe that the Dating category app have the highest follow by Medical category. This is because the dating app tends to attract more teen which have higher income. Therefore the average price for dating category app have higher price.

### Q5: Which premium app category have the most install record?

This question required groupby() function which allow us to group the category together.

```python
## We will use back the previous question code then groupby category
premium_app_install = premium_app.groupby(['Category']).count()

## Next we calculate the percentage according to each category
premium_app_install['Percentage'] = (premium_app_install['App Name'] / premium_app_install['App Name'].sum()) * 100
premium_app_install = premium_app_install.drop(['App Name', 'App Id', 'Rating', 'Rating Count',
       'Maximum Installs', 'Free', 'Price', 'Size', 'Minimum Android',
       'Developer Id', 'Released', 'Last Updated', 'Content Rating'], axis = 1)
```

```python
premium_app_install.plot(kind='bar', figsize=(10,5), color = 'red')
plt.title("Premium Apps category with highest percentage install")
plt.ylabel('Number of Installs')
plt.xlabel('Category')
```

From the bar chart we can observe that the Education category is the most popular among the premium apps follow by Personalization and Tools. <br>
The reason is because education have become very common in our world in each stage of mankind.
