**<h2>Introduction to Koalas<h2>**

---

The Koalas project makes data scientists more productive when interacting with big data, by implementing the pandas DataFrame API on top of Apache Spark.

pandas is the de facto standard (single-node) DataFrame implementation in Python, while Spark is the de facto standard for big data processing. With this package, you can:


1.   Be immediately productive with Spark, with no learning curve, if you are already familiar with pandas.
2.   Have a single codebase that works both with pandas (tests, smaller datasets) and with Spark (distributed datasets).

```python
import pandas as pd
import numpy as np
import databricks.koalas as ks
```

```python
#Read data.csv file using koalas
%time kdf = ks.read_csv('yellow_tripdata_2015-01.csv')
```

**<h2>Viewing Data<h2>**

---

```python
#See the top rows of the frame
%time kdf.head()
```

```python
#Describe shows a quick statistic summary of your data

%time kdf.describe()
```

**<h2>Missing Data<h2>**

---

```python
#To drop any rows that have missing data.
%time kdf.dropna()
```

**<h2>Operations<h2>**

---

```python
#Compute mean of each numerical column using Koalas
%time kdf.mean(axis=0)
```

**<h2>Grouping<h2>**

---

```python
#Grouping and then applying the sum() function to the resulting groups.
%time kdf.groupby('RateCodeID').sum()
```

**<h2>Parquet<h2>**

---

```python
#write a Parquet file
%time kdf.to_parquet('bar.parquet')
```

```python
#read a Parquet file
%time ks.read_parquet('bar.parquet').head(10)
```

**<h2>Spark IO<h2>**

---
Koalas fully support Spark’s various datasources such as ORC and an external datasource.

```python
# write it to the specified datasource and here to read it from the datasource.
%time kdf.to_spark_io('zoo.orc', format="orc")
```

```python
%time ks.read_spark_io('zoo.orc', format="orc").head(10)
```
