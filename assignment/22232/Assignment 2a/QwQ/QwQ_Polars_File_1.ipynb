##Expressions

```python
polars_df.select(pl.col('*'))
```

To select a column we need to do two things. Define the DataFrame we want the data from. And second, select the data that we need. In the example above, we select col('*'). The asteriks stands for all columns.

```python
polars_df.select([pl.col('StateCode'), pl.col('BusinessYear')]).limit(5)
```

We can also specify the specific columns that we want to return. There are two ways to do this. The first option is to create a list of column names. Second, specify each column within a list in the select statement as shown above.

```python
polars_df.select([pl.exclude('Couple')])
```

We can also exclude the entire column from our view by using exclude in select statement.

```python
polars_df.filter(pl.col("IndividualRate").is_between(float(29.0), float(32.0)),)
```

The filter option allows us to create a subset of the DataFrame.

```python
polars_df.groupby(["StateCode"], maintain_order=True).agg([pl.col("*").count()])
```

We can create a new dataframe by using groupby function. As shown above, the groupby method splits the DataFrame into groups based on the values in the specified column(s), and the agg function applies a function to each group and returns a new DataFrame with the aggregated results.

In this case, the pl.col("*").count() expression is used as the aggregation function, which counts the number of rows in each group. The maintain_order=True argument specifies that the order of the rows in the original DataFrame should be preserved in the resulting DataFrame.

```python
polars_df.with_column(pl.col("IndividualTobaccoRate").fill_null(pl.lit('Unknown'),),)
```

To fill the null column, we can use the above code to fill with a value that we want.

```python
polars_df.select(pl.pearson_corr("*", "*"))
```

Polars also can do some calculation of the Pearson correlation coefficient as above. And also Spearman correlation.

**Aggregation**

```python
polars_df.groupby("BusinessYear", maintain_order=True).agg(pl.col("Age").agg_groups())
```

The agg method is being applied to the resulting grouped data. It appears to be using the agg_groups function from the col method of the plotly library (aliased as pl) as an aggregation function. The agg_groups function groups the data based on the values in a specified column and returns the resulting groups as a list. In this case, it looks like the column being used to group the data is "Age".
