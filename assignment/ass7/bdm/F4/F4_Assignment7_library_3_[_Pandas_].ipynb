## Join

In this dataframe, there’s an overlapping column named ‘Temperature’, it will be renamed to ‘Temperature_df’ in the original DataFrame df and ‘Temperature_df_temp’ in the DataFrame df_temp.

```python
%%time
df_joined = df.join(df_temp, on='Beer_Style', lsuffix='_df', rsuffix='_df_temp')
```

In this dataset, the overlapping column ‘Temperature’ will be renamed to ‘Temperature_x’ in the original DataFrame df and ‘Temperature_y’ in the DataFrame df_temp.

In the context of pandas’ merge() function, how='left' specifies that a left join should be performed.

A left join returns all the rows from the left DataFrame and the matched rows from the right DataFrame. If there is no match, the result is NaN in the columns of the right DataFrame.

```python
%%time
df_joined2 = df.merge(df_temp, on='Beer_Style', how='left')
df_joined2
```

## EDA Visualization

Exploratory Data Analysis (EDA) is a crucial step in data science that involves analyzing and investigating datasets to summarize their main characteristics, often with visual methods. You can create various plots like histograms, box plots, scatter plots, and more using Matplotlib and Seaborn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
```

### Barplot

```python
%%time

Beer_count = df.Beer_Style.value_counts().rename_axis('Beer').reset_index(name='count')
Beer_count
Beer_count['count'] = pd.to_numeric(Beer_count['count'])
print(Beer_count.dtypes)
```

```python
%%time

# Plot the DataFrame
Beer_count.plot(kind='bar', x='Beer', y='count', color=['red', 'orange', 'yellow', 'green', 'cyan', 'blue', 'indigo','purple'])
plt.xlabel('Beer')
plt.ylabel('Count')
plt.title('Bar Plot for Beer Count')
plt.show()

```

### Boxplot

```python
%%time

df.boxplot(column='Quality_Score')
plt.show()
```

### Scatterplot


```python
%%time

## Reduce the size of data by applying Selection Quality Score Above mean score > 6

selected_df = df[df["Quality_Score"] > 5]

x = selected_df["Bitterness"]
y = selected_df["Color"]

plt.figure(figsize=(12,6))
plt.scatter(x, y, c="red", alpha=0.5)
plt.xlabel('Bitterness')
plt.ylabel('Color')
plt.show()
```

### Lineplot

```python
%%time

df['Brew_Date'] = pd.to_datetime(df['Brew_Date'])

df['year'] = df['Brew_Date'].dt.year
df['month'] = df['Brew_Date'].dt.month
df['day'] = df['Brew_Date'].dt.day
```

```python
%%time

df_filtered = df[((df.year==2023))&(df.month==1)&((df.Location=="HSR Layout"))&((df.Beer_Style=="Wheat Beer"))&((df.SKU=="Bottles"))&(df.Quality_Score>5)]
df_filtered.head(3)
```

```python
%%time

# Plot the historical Sales
plt.figure(figsize=(30,10))
plt.plot(df_filtered.Brew_Date, df_filtered.Total_Sales)
plt.xlabel('Brew_Date')
plt.ylabel('Sales')
plt.title('2023 January Historical Sales of Wheat Beer(Bottles) with Quality Score 5 and above at HSR Layout')
plt.show()
```

### What is the distribution of Alcohol Content in the dataset?

### Is there a relationship between pH Level and Bitterness in the brewing process?

```python
%%time

plt.figure(figsize=(10,6))
sns.scatterplot(x='pH_Level', y='Bitterness', data=df)
plt.title('pH Level vs Bitterness')
plt.show()
```

### What are the top 5 Beer Styles based on Total Sales?

```python
%%time

top_beer_styles = df.groupby('Beer_Style')['Total_Sales'].sum().sort_values(ascending=False).head(5)
print("Top 5 Beer Styles based on Total Sales:\n", top_beer_styles)
```

```python
%%time

# Plot the DataFrame
top_beer_styles.plot(kind='bar', x='Beer_Style', y='Total_sales', color=['red', 'orange', 'yellow', 'green', 'cyan', 'blue', 'indigo','purple'])
plt.xlabel('Beer')
plt.ylabel('Count')
plt.title('Beer Rank')
plt.show()
```

### Which Location produced the most beer?

```python
%%time

top_location = df.groupby('Location')['Volume_Produced'].sum().idxmax()
print("Location that produced the most beer: ", top_location)
```

### What is the distribution of Fermentation Time, and are there any outliers?

```python
%%time

plt.figure(figsize=(10,6))
sns.boxplot(x=df['Fermentation_Time'])
plt.title('Distribution of Fermentation Time')
plt.show()
```

### What is the Correlation between Alcohol Content and Quality Score?

```python
%%time

correlation = df['Alcohol_Content'].corr(df['Quality_Score'])
print("Correlation between Alcohol Content and Quality Score: ", correlation)
```
