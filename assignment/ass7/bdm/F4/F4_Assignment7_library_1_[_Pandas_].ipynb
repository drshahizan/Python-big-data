# Pandas Installation

Pandas is a Python library that simplifies data analysis. It offers two main structures: Series for single-column data and DataFrame for multi-column data.


With Pandas, data can be clean by fixing errors or filling in missing values. It allows you to analyze data by calculating statistics, grouping data, and finding patterns. Pandas can be used alongside other libraries like NumPy, SciPy, Matplotlib, and Scikit-learn for additional functionalities such as mathematical computations, scientific analysis, plotting, and machine learning.


It also enables to read data from and write data to various file formats or databases. Furthermore, it can be used with Matplotlib to create visualizations. Overall, Pandas is a vital tool for data scientists and data analysts as it makes data analysis in Python more efficient.

## Group-by

```python
# Group by "Beer Style" and calculate mean and std for "Fermentation_Time" & "Temperature"
%%time
df_group = df.groupby(by='Beer_Style').agg({'Fermentation_Time':['mean','std'],'Temperature':['mean','std']})
df_group
```

```python
# Group by "Beer_Style" and calculate mean for "Temperature"
%%time
df_temp = df.groupby('Beer_Style')['Temperature'].mean()
df_temp
```

```python
# Group by "Beer_Style" and calculate count
%%time
df_count = df.groupby('Beer_Style').count()
df_count
```

## Join

```python
# Pandas default join type is inner join
%%time
df_joined = pd.merge(df, df_group, on='Beer_Style')
df_joined
```

## EDA Visualization

Exploratory Data Analysis (EDA) is a crucial step in data science that involves analyzing and investigating datasets to summarize their main characteristics, often with visual methods. You can create various plots like histograms, box plots, scatter plots, and more using Matplotlib and Seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Barplot

```python
%%time

Beer_count = df.Beer_Style.value_counts().rename_axis('Beer').reset_index(name='count')
Beer_count
```

```python
# Show boxplot, only use this part for execution time calculation

%%time
Beer_count.plot(kind='bar', x='Beer', y='count', color=['red', 'orange', 'yellow', 'green', 'cyan', 'blue', 'indigo','purple'])
plt.xlabel('Beer')
plt.ylabel('Count')
plt.title('Bar Plot for Beer Count')
plt.show()
```

### Boxplot

```python
%%time

df.boxplot(column='Quality_Score')
plt.show()
```

### Scatterplot


```python
%%time
df.Quality_Score.mean()
```

```python
%%time
# Reduce the size of data by applying Selection Quality Score Above mean score > 8
selected_df = df[df["Quality_Score"] > 8]
x = selected_df["Bitterness"]
y = selected_df["Color"]
```

```python
%%time
# Show Scatterplot, only use this part for execution time calculation
plt.figure(figsize=(12,6))
plt.scatter(x, y, c="red", alpha=0.5)
plt.xlabel('Bitterness')
plt.ylabel('Color')
plt.show()
```

### Lineplot

```python
%%time

df['Brew_Date'] = pd.to_datetime(df['Brew_Date'])
df['year'] = df['Brew_Date'].dt.year
df['month'] = df['Brew_Date'].dt.month
df['day'] = df['Brew_Date'].dt.day
```

```python
%%time

df_filtered = df[((df.year==2023))&(df.month==1)&((df.Location=="HSR Layout"))&((df.Beer_Style=="Wheat Beer"))&((df.SKU=="Bottles"))&(df.Quality_Score>8)]
df_filtered.head(3)
```

```python
%%time

# Show lineplot, only use this part for execution time calculation
plt.figure(figsize=(30,10))
plt.plot(df_filtered.Brew_Date, df_filtered.Total_Sales)
plt.xlabel('Brew_Date')
plt.ylabel('Sales')
plt.title('2023 January Historical Sales of Wheat Beer(Bottles) with Quality Score 8 and above at HSR Layout')
plt.show()
```

### Is there a relationship between pH Level and Bitterness in the brewing process?

```python
%%time

plt.figure(figsize=(10,6))
sns.scatterplot(x='pH_Level', y='Bitterness', data=df)
plt.title('pH Level vs Bitterness')
plt.show()
```

### What are the top 5 Beer Styles based on Total Sales?

```python
%%time

top_beer_styles = df.groupby('Beer_Style')['Total_Sales'].sum().sort_values(ascending=False).head(5)
print("Top 5 Beer Styles based on Total Sales:\n", top_beer_styles)
```

```python
%%time

# Plot the DataFrame
top_beer_styles.plot(kind='bar', x='Beer_Style', y='Total_sales', color=['red', 'orange', 'yellow', 'green', 'cyan', 'blue', 'indigo','purple'])
plt.xlabel('Beer')
plt.ylabel('Count')
plt.title('Beer Rank')
plt.show()
```
