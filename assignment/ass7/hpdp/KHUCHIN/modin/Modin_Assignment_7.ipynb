### 12. Replacing string column missing value with the calculated mode.

```python
# Specify the string columns in the subset
string_cols = df_2017.select_dtypes(include='object').columns

# Replace missing values with mode for string columns
for col in string_cols:
    mode_value = df_2017[col].mode()[0]  # Calculate mode (assumes there's a unique mode)
    df_2017[col].fillna(mode_value, inplace=True)
```

### 20. Ask and answer questions about the DataFrame
##### Question 1: What is the distribution of flight delays?

```python
import modin.pandas as pd
import matplotlib.pyplot as plt

# Convert Modin DataFrame to Pandas DataFrame
df_2017_pandas = pd.DataFrame(df_2017)

# Plot histogram of departure delays using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df_2017_pandas['DEP_DELAY'].dropna(), bins=30, density=True, alpha=0.7)
plt.title('Distribution of Departure Delays')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Density')
plt.show()
```

##### Question 2: Which airlines have the most and least delays?

```python
# Group by airline and calculate average departure delay
average_delay_by_airline = df_2017.groupby('OP_CARRIER')['DEP_DELAY'].mean().sort_values()

# Plot bar chart
plt.figure(figsize=(12, 8))
average_delay_by_airline.plot(kind='barh', color='skyblue')
plt.title('Average Departure Delay by Airline')
plt.xlabel('Average Departure Delay (minutes)')
plt.ylabel('Airline')
plt.show()
```

###### - Airline that has the most delay is B6 while the least delay is HA

##### Question 3: What is the relationship between distance and air time?

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='DISTANCE', y='AIR_TIME', data=df_2017, alpha=0.5)
plt.title('Relationship between Distance and Air Time')
plt.xlabel('Distance (miles)')
plt.ylabel('Air Time (minutes)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hexbin(df_2017['DISTANCE'], df_2017['AIR_TIME'], gridsize=20, cmap='viridis')
plt.title('Relationship between Distance and Air Time (Hexbin)')
plt.xlabel('Distance (miles)')
plt.ylabel('Air Time (minutes)')
plt.colorbar(label='Count')
plt.show()
```

###### - Shorter the distance, shorter the air time

##### Question 4: Are there seasonal trends in flight cancellations?

```python
import calendar

# Convert Modin DataFrame to Pandas DataFrame
df_2017_pandas = pd.DataFrame(df_2017)

# Group by month and calculate cancellation percentage
monthly_cancellation_percentage = (
    df_2017_pandas.groupby(df_2017_pandas['FL_DATE'].dt.month)['CANCELLED'].mean() * 100
)

# Plot line chart of cancellation percentage by month
plt.figure(figsize=(10, 6))
plt.plot(monthly_cancellation_percentage.index, monthly_cancellation_percentage.values, marker='o')
plt.title('Cancellation Percentage by Month')
plt.xlabel('Month')
plt.ylabel('Cancellation Percentage')
plt.xticks(range(1, 13), calendar.month_abbr[1:])
plt.grid(True)
plt.show()
```

###### - The increasing trend start from May to Sep and decrease a lot in October.

##### Question 5: Which day of the week has the highest average departure delay?

```python
import datetime

# Convert 'FL_DATE' to datetime
df_2017['FL_DATE'] = pd.to_datetime(df_2017['FL_DATE'])

# Create a new column for the day of the week (0 = Monday, 6 = Sunday)
df_2017['DayOfWeek'] = df_2017['FL_DATE'].dt.dayofweek

# Group by day of the week and calculate average delay
average_delay_by_day = df_2017.groupby('DayOfWeek')['DEP_DELAY'].mean()

# Define day names
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Plot line chart with labeled days
plt.figure(figsize=(10, 6))
plt.plot(day_names, average_delay_by_day, marker='o', color='salmon', linestyle='-', linewidth=2)
plt.title('Average Departure Delay by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Departure Delay (minutes)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.grid(True)
plt.show()
```

###### - Friday is the highest average departure delay
